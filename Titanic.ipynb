# 4.Find out missing values and replace them for the applicable columns.

```python
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].fillna('unknown')
df.isnull().sum()
```

```python
df['Embarked']=df['Embarked'].fillna(df.mode)
df.isnull().sum()
```

```python
C=df.select_dtypes(include=['object','category']).columns
C
```

```python
d=pd.get_dummies(df,columns=['Sex'],drop_first=True)
d
```
